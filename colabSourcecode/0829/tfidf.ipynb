```python
import pandas as pd
path=pd.read_csv('combinedSentences.csv')
path.drop('1',axis=1,inplace=True)
path.head(5)
```

```python
sentence=path.values
sentence
```

```python
flat_list = [item[0] for item in sentence]
print(flat_list)
```

```python
seg_list = [jieba.cut(text) for text in flat_list]

# 將分詞結果轉換為所需格式
segmented_list = [' '.join(words) for words in seg_list]

# 打印結果
print(segmented_list)
```

```python
vectorizer=CountVectorizer(stop_words=None, token_pattern="(?u)\\b\\w+\\b")
X=vectorizer.fit_transform(segmented_list)
r=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
print('CountVector')
r
```

```python
transformer = TfidfTransformer(smooth_idf=True)
Z = transformer.fit_transform(X)
r = pd.DataFrame(Z.toarray(),columns=vectorizer.get_feature_names_out())
print("TFIDF")
r
```

```python
print("IDF")
pd.DataFrame([transformer.idf_], columns=vectorizer.get_feature_names_out())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

k = range(1,200)
sum_squared_errors = []

for i in k:
  model = KMeans(n_clusters=i,init='k-means++',n_init='auto')
  model.fit_predict(Z)
  sum_squared_errors.append(model.inertia_)

plt.plot(k,sum_squared_errors)
plt.xlabel('K-Value')
plt.ylabel('Sum of Squared Errors')
```
